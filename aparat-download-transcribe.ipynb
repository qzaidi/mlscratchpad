The following code uses huggingface transformers - if the model is not already there, it would be downloaded to ~/.cache/huggingface/hub/

```python
import torch
from transformers import pipeline

pipe = pipeline("automatic-speech-recognition","openai/whisper-large-v3",torch_dtype=torch.float16,device="cuda:0")
```

```python
%%time
res = pipe("./outputfile.mp4", generate_kwargs={"language": "fa"})
```

```python
print(res)
```

We are forced to use whisper and load audio first, because it seems transformers doesnt accept mp4 videos, it will only accept audio. 
Also if you get an error about cuda memory, you will need to reset cuda cache.

```python
import whisper
audio = whisper.load_audio('./urdu-ns.mp4')
pipe = pipeline("automatic-speech-recognition","openai/whisper-large-v3",torch_dtype=torch.float16,device="cuda:0")
resns = pipe(audio,generate_kwargs={"language": "<|ur|>"})
```

```python
print(resns)
```

```python
pipe(audio,generate_kwargs={"language": "<|ur|>","task":"translate"})
```
